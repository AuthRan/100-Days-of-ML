```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.sample(10)
```

```python
titanic = pd.read_csv('data/train.csv')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

Ab for bivariate analysis you can have several cases which are discussed below <br>
I.Numerical-Numerical : <b>Scatter Plot</b>

```python
# sns.scatterplot(tips['total_bill'], tips['tip'])  -> will throw an error
sns.scatterplot(x='total_bill', y='tip', data=tips); 
# to remove <Axes: xlabel='total_bill', ylabel='tip'> line u should use either semicolon at the end or use plt.show()
```

```python
# now if you want to do mutlivariate analysis on this same thing like batao inme se kon male hai ya female ie you are analyzing three attributes total_bill, tip and sex you should use 
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex');
```

```python
# Now if you want to go fully multivariate then
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker', size='size');
```

II.Bar Plot(Numerical-Categorical)

```python
titanic.sample(5)
```

```python
# Suppose you waana see the relationship btw Pclass and Average age traveling in it
# Usually x axis pe categorical values hoti hai and y axis pe numerical
sns.barplot(x='Pclass', y='Age', data=titanic)
```

```python
# If you want isme bhi u can do multivariate analysis like avg age gender wise in each Pclass
sns.barplot(x='Pclass', y='Age', data=titanic, hue='Sex')
```

III.Box Plot(Numerical-Categorical)

```python
# You can plot box plot of multiple attributes sathme using seaborn like this suppose you want to know relation btw sex and age along with who survived
# sns.boxplot(x='Sex', y='Age', data=titanic) -> basic
sns.boxplot(x='Sex', y='Age', data=titanic, hue='Survived')
# the graph tells us that in male younger population survived but in female older population survived can you tell why ? hehe
```

IV.Distplot(Categorical-Numerical)

```python
# sns.distplot(titanic['Age']);
sns.distplot(titanic[titanic['Survived']==0]['Age'], hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'], hist=False)
```

V.HeatMap(Categorical-Categorical)

```python
titanic.head(5)
```

```python
heatmap_data = pd.crosstab(titanic['Pclass'], titanic['Survived'])
heatmap_data
```

```python
# Ab aap iss data ka bhi graph plot kr skte ho called heatmap
# sns.heatmap(heatmap_data) -> basic way to plot heatmap
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm')
```

```python
# titanic.groupby('Pclass').mean()['Survived']*100   -> error bcz of non numeric column present
# titanic.groupby('Pclass', numeric_only=True).mean()['Survived'] * 100 -> error
titanic.groupby('Pclass').mean(numeric_only=True)['Survived'] * 100
```

VI.ClusterMap(Categorical-Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
# It tells you about konsa data kitna similar hai ek dusre ke
```

VII. Pair Plot(Numerical-Numerical + Sabka)

```python
# Suppose apke pass multiple numerical data attributes hai aur apko multiple data attributes ke sath scatter plot relationship dikhana hai then instead of ek ek ka scatter plot banana you can make sabka ek sath using pairplot
# Yeh apne aap numerical data ko pkdke scatter plot bana dega aur khudka sath histogram plot krdega
iris
```

```python
sns.pairplot(iris)
```

```python
# Now you can pass one more parameter hue if you want to know konsa dot akhir konsa species ko repersent kr rha hai
sns.pairplot(iris, hue='species')
```

VIII. LinePlot (Numerical-Numerical)

```python
# It is a special case of scatter plot ie scatter plot ke sare lines ko agr aap jod do then its a line plot 
# It is preffered over scatter plot when your x axis has time data
flights 
```

```python
new = flights.groupby('year')['passengers'].sum()
new.head()
```

```python
sns.lineplot(new)
```
