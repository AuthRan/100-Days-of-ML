```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/train.csv')
df.sample(5)
```

<h1>Categorical Data</h1>
<h3>(a)Count Plot</h3>

```python
# Univariate analysis on a categorial data 
# Lets first see survived column - as it is useful column
sns.countplot(x='Survived', data=df)
# df['Survived'].value_counts().plot(kind='bar')
```

```python
# Similarly you can also plot Pclass also 
sns.countplot(x='Pclass', data=df)
```

(b)Pie Chart

```python
# You can also make pie charts of data given its very simple just always wherever u are rem this code : df['Survived'].value_counts().plot(kind='bar')
df['Survived'].value_counts().plot(kind='pie')
```

```python
# Ab iss upr wale plot mein agr percentage dikhana ho toh do this
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
df['Embarked'].value_counts().plot(kind='pie', autopct='%.2f')
```

<h1>Numerical Data</h1>
<h3>(a)Histogram</h3>

```python
# Now categorical mein toh asan tha pie ya bar se dikhado konsa category kitna aaya hai
# In case of numerical you may use min max mode and all but before that can you make numerical data behave as categorical data? 
# Yes! Histogram ie ki numerical data ko range mein yani bins mein break krdete ho.
plt.hist(df['Age'])
# sns.histplot(df['Age'])
```

<h3>(b)Distogram</h3>

```python
# It is an improvement of histogram 
# Along with histogram it give kde kerner distribution estimation isko hum pdf yani probability desity function ie y axis pe probability and x pe hoga age for example for any x lets say 40 then uss age ka passenget hone ki probability btayega y axix whearas histogram gives u exact value. You will understand its importance when u read multivariate analysis.
sns.distplot(df['Age'])
```

<h3>(c)BoxPlot</h3>
<img src="images/boxplot.png" alt="Boxplot Explanation" width="600">


```python
# It is used to give 5 number summary
# Median - After sorting all the datapoints Bich wala value, Q1:25th percentile, Q3:25th percentile, Max/Min:Q1+1.5IQR(inter quartile range). Aur iske bhar jo bhi ho woh potential outlier hota hai.
# So in future if you work with noisy data then you can plot a boxplot on a column to know how much noisy it is actually.
sns.boxplot(df['Age'])
```

```python
# You can now do normal math function on any column as
df['Age'].mean()
```

```python
df['Age'].median() # which matches with the graph
#df['Age].max()
```

```python
# You can calculate skewness of graph with skew function. SKewness of a graph means ki koi graph kitna symmetrical hai yani jitna max value utna hi min value bhi ho.
# Skew() 0 mean perfectly skewed aur jitna positive mtlb utna positively skewed towards right side and if negative hai then negatively skewed towards left side.
df['Age'].skew()
```
